```python
class HollowNode():
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.child = None
        self.next = None
        self.ep = None
        self.rank = 0
        

    
```

```python
class HollowHeap():
    def __init__(self):
        self.vals = {} # store val, node
        print("initiated")
        
        
    def insert(self, v, k, h):
        return self.meld(self.makeNode(k, v), h)
    
    def meld(self, g, h):
        if g == None:
            return h
        if h == None:
            return g
        return self.link(g, h)
        
    def findMin(self, h):
        if h == None:
            return None
        return h.item
    
    def decKey(self, k, v, h):
        
        if (k >= self.vals[v].key):
            print("key is greater, returning")
            return
        
        # if root
        u = self.vals[v]
        if u == h:
            u.key = k
            return h
        # otherwise
        v = self.makeNode(k, v)
        u.item = None
        if u.rank > 2:
            v.rank = u.rank - 2
        v.child = u
        u.ep = v
        
        return self.link(v, h)
        
    def deleteMin(self, h):
        return self.delete(h.value, h)
    
    def delete(self, v, h):
        self.vals[v].item = None
        self.vals[v] = None
        # not root del
        if h.item != None:
            return h
        maxRank = 0
        A = [None] * 1000
        
        # otherwise
        while h != None:
            w = h.child
            v = h
            h = h.next
            while w != None:
                u = w
                w = w.next
                if u.item == None:
                    if u.ep == None:
                        u.next = h
                        h = u
                    else:
                        if u.ep == v:
                            w = None
                        else:
                            u.next = None
                        u.ep = None
                else:
                    # do ranked links
                    while A[u.rank] != None:
                        u = self.link(u, A[u.rank])
                        A[u.rank] = None
                        u.rank += 1
                    A[u.rank] = u
                    if u.rank > maxRank:
                        maxRank = u.rank
                        
        # do unranked links
        for i in range(maxRank + 1):
            if A[i] != None:
                if h == None:
                    h = A[i]
                else:
                    h = self.link(h, A[i])                
        return h
    # aux methods
    
    def makeNode(self, k, v):
        u = HollowNode(k, v)
        self.vals[v] = u
        return u
        
    @staticmethod
    def link(v, w):
        if v.key >= w.key:
            self.addChild(v, w)
            return w
        else:
            self.addChild(w, v)
            return v
                
    @staticmethod
    def addChild(v, w):
        v.next = w.child
        w.child = v
        
```

```python
heap = HollowHeap()
h = heap.insert(1, 3, None)
h = heap.deleteMin(h)
```
