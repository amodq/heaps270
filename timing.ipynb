```python
import time
from hollow import *
import random
def TestHollowRandom(n):
    heap = HollowHeap()
    h = None
    vals = list(range(2, 3 + n))
    random.shuffle(vals)
    t = time.process_time()
    nodes = []
    for i in range(n):
        k = random.random()
        a = random.random()

        if a < 0.3:
            h = heap.insert(k, vals[i], h)
            nodes.append(vals[i])
        elif a < .5:
            if h == None or not nodes:
                continue
            h = heap.decKey(k, random.choice(nodes), h)
        else:
            if h == None:
                continue
            nodes.remove(h.value)
            h = heap.deleteMin(h)
    t1 = time.process_time() - t
    return t1
```

```python
import time
from fib import *


def TestFibRandom(n):
    heap = FibonacciHeap()
    nodes = []
    t = time.process_time()
    for i in range(n):
        k = random.random()
        a = random.random()
        if a < 0.33:
            h = heap.insert(k)
            nodes.append(h)
        elif a < .5:
            if not nodes:
                continue
            update = random.choice(nodes)
            heap.decrease_key(update, k)
        else:
            x = heap.extract_minimum()
            if x:
                nodes.remove(x)
    t1 = time.process_time() - t
    return t1
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = []
fib = []
hollow = []
#binom = []

for i in range(6):
    n = 5 ** (i + 5)
    x.append(np.log(n))
    fib.append(TestFibRandom(n))
    hollow.append(TestHollowRandom(n))
    
plt.plot(x,\
         fib, label='Fib')
# plt.plot(x,\
#          binom, label='Binom')

plt.title('Runtime on Random Operations')
plt.plot(x,\
         hollow, label='Hollow')

plt.legend()
plt.xlabel('Log Size of Operations')
plt.ylabel('Runtime (s)')
```
