```python
class HollowNode():
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.child = None
        self.next = None
        self.ep = None
        self.rank = 0
        

    
```

```python
class HollowHeap():
    def __init__(self):
        self.vals = {} # store val, node
        print("initiated")
        
        
    def insert(self, k, v, h):
        return self.meld(self.makeNode(k, v), h)
    
    def meld(self, g, h):
        if g == None:
            return h
        if h == None:
            return g
        return self.link(g, h)
        
    def findMin(self, h):
        if h == None:
            return None
        return h.item
    
    def decKey(self, k, v, h):
        
        if v not in self.vals:
            return h
        
        if (k >= self.vals[v].key):
            return h
        
        # if root
        u = self.vals[v]
        if u == h:
            u.key = k
            return h
        # otherwise
        v = self.makeNode(k, v)
        u.item = None
        if u.rank > 2:
            v.rank = u.rank - 2
        v.child = u
        u.ep = v
        
        return self.link(v, h)
        
    def deleteMin(self, h):
        return self.delete(h.value, h)
    
    def delete(self, v, h):
        self.vals[v].item = None
        self.vals[v] = None
        # not root del
        if h != None:
            return h
        maxRank = 0
        A = [None] * 10000
        
        # otherwise
        while h != None:
            w = h.child
            v = h
            h = h.next
            while w != None:
                u = w
                w = w.next
                if u.item == None:
                    if u.ep == None:
                        u.next = h
                        h = u
                    else:
                        if u.ep == v:
                            w = None
                        else:
                            u.next = None
                        u.ep = None
                else:
                    # do ranked links
                    while A[u.rank] != None:
                        u = self.link(u, A[u.rank])
                        A[u.rank] = None
                        u.rank += 1
                    A[u.rank] = u
                    if u.rank > maxRank:
                        maxRank = u.rank
                        
        # do unranked links
        for i in range(maxRank + 1):
            if A[i] != None:
                if h == None:
                    h = A[i]
                else:
                    h = self.link(h, A[i])                
        return h
    # aux methods
    
    def makeNode(self, k, v):
        u = HollowNode(k, v)
        self.vals[v] = u
        return u
        
    def link(self, v, w):
        if v.key >= w.key:
            self.addChild(v, w)
            return w
        else:
            self.addChild(w, v)
            return v
                
    @staticmethod
    def addChild(v, w):
        v.next = w.child
        w.child = v
        
```

```python
import random
from heapq import *
from hollow import *

correct = []
heap = HollowHeap()
h = heap.insert(.5, 3, None)
heappush(correct, (.5, 3))

actions = list(range(2, 1005))
random.shuffle(actions)

for i in range(1000):
    k = random.random()
    a = random.random()
    
    if a < 0.33:
        heappush(correct, (k, actions[i]))
        h = heap.insert(k, actions[i], h)
        if not (correct[0][0] == h.key and correct[0][1] == h.value):
            print("error!")
    elif a < .9:
        if h == None or i == 0:
            continue
        update = random.choice(actions[0:i])
        for j in range(len(correct)):
            if correct[j][1] == update:
                if correct[j][0] > k:
                    correct[j] = (k, update)
                    break
        correct.sort(key = lambda x: x[0])
        h = heap.decKey(k, update, h)
    else:
        if not h:
            continue
        h = heap.deleteMin(h)
        heappop(correct)
```

```python
import time
from hollow import *
import random
def TestHollowRandom(n):
    heap = HollowHeap()
    h = None
    vals = list(range(2, 3 + n))
    random.shuffle(vals)
    t = time.process_time()
    nodes = []
    for i in range(n):
        k = random.random()
        a = random.random()

        if a < 0.3:
            h = heap.insert(k, vals[i], h)
            nodes.append(vals[i])
        elif a < .5:
            if h == None or i == 0 or not nodes:
                continue
            h = heap.decKey(k, random.choice(nodes), h)
            #print(h.value, nodes)
        else:
            if h == None or i == 0:
                continue
            #print(h.value, nodes)
            nodes.remove(h.value)
            h = heap.deleteMin(h)
    t1 = time.process_time() - t
    return t1
```

```python
import time
from fib import *
def TestFibRandom(n):
    heap = FibonacciHeap()
    h = None
    nodes = []
    t = time.process_time()
    for i in range(n):
        k = random.random()
        a = random.random()

        if a < 0.6:
            print("append")
            h = nodes.append(heap.insert(k))
        elif a < .9:
            print("decrease key")
            if heap == None or i == 0 or not nodes:
                continue
            h = heap.decrease_key(random.choice(nodes), k)
        else:
            print("extract min")
            if heap == None or i == 0:
                continue
            h = heap.extract_minimum()
    t1 = time.process_time() - t
    return t1
```

```python
import time
from binom import *
def TestBinomRandom(n):
    heap = BinomialHeap()
    h = None
    vals = list(range(2, 3 + n))
    random.shuffle(vals)
    t = time.process_time()
    nodes = []
    for i in range(n):
        k = random.random()
        a = random.random()

        if a < 0.6:
            nodes.append(heap.insert(k, vals[i]))
        elif a < .9:
            if i == 0 or not nodes:
                continue
            update = random.choice(nodes)
            heap.dec_key(update, k)
        else:
            heap.del_min()
    t1 = time.process_time() - t
    return t1
```

```python
from hollow import *
def DijkstraTest():
    import time
    u = random.randint(0, 100)
    v = random.randint(0, 100)
    
    t = time.process_time()
    
    heap = HollowHeap()
    h = None
    h = heap.insert(u, 0, h)
    
    distances = {}
    distances[u] = 0
    
    while h:
        n = h.key
        d = h.value
        if n == v:
            print("found v with distance", d)
            break
        h = heap.deleteMin(h)
        edges = list(G.edges(n))
        
        for e in edges:
            newd = d + G.edges[e[0], e[1]]['weight']
            if e[1] not in distances or newd < distances[e[1]]:
                distances[e[1]] = newd
                h = heap.insert(e[1], newd, h)
    
    t1 = time.process_time()
    return t1 - t
```
